# Solenoids

One of the fundamental components of an electric circuit is the inductor.  An inductor is typically formed by wrapping a conductor (electrical wire) around a core (usually air or an iron cylinder).  The shaped formed by this winding fits the concept of a **solenoid**.

## Generic Example

Suppose we have a solenoid centered around the y-axis of a $x$, $y$, $z$ Cartesian space with the base unit of each axis being meters.  This solenoid has a radius of $r$ meters and a "psuedo-frequency" of $\mathrm{pf}$ turns (cycles) per meter.  The angular "velocity" ($w$) in radians is given by:  $ w = 2\pi \mathrm{pf} $.

From these presuppositions, it follows that:  

$$ x^2 + z^2 = r^2$$

$$ z = r \cdot \sin \theta$$

and,
$$ x = r \cdot \cos \theta$$

where $\theta:=\theta(y) = w y$.

The above equations give our solenoid.

Now suppose there exists a variable vector in a 3D Cartesian vector space, where every unit vector has a magnitude of one meter -- the same space that our solenoid rests in -- that is equal to the sum of of two other variable vectors, denoted as '$\mathbf{y}$' and '$\mathbf{r}$', whose terminal point always rests on our solenoid.

$$ \mathbf{y} := \mathbf{y}(t) = V_y\, t\cdot \mathbf{j}$$

$$ \mathbf{r} := \mathbf{r}(t) = r \,\angle\, \theta(t)$$

Basically, $\mathbf{y}$ has a magnitude that varies with time ($t$) and a direction that is constant, always pointing in the direction of the positive end of the y-axis.  The other vector, $\mathbf{r}$, has a constant magnitude of $r$, but a direction that varies with time and subsequently, the magnitude of $\mathbf{y}$.

For $\theta = \theta(\mathbf{y}_y) = \theta(V_y \, t) = \theta (t)$ to hold true, $\mathbf{r}$ must rotate around the y-axis with a frequency, denoted as '$f$', such that:  

$$ \theta(t) = \omega t = 2\pi f t = w \mathbf{y}_y = 2\pi \mathrm{pf} \mathbf{y}_y  = \theta(y) $$

Dividing throughout the above equation by $2\pi$ gives:  

$$ f t = \mathrm{pf} \mathbf{y}_y= \mathrm{pf} V_y \, t$$

Dividing throughout by $t$ gives:  

$$ f = \mathrm{pf}\,V_y$$

Since the frequency is in terms of cycles per second, and the *psuedo*-frequency is in terms of turns per meter -- which corresponds to cycles per meter -- $f$ over $\mathrm{pf}$  is a quantity that is in terms of meters per second, which is what $y$ over $t$ is in terms of and is unit of measurement of the velocity of $\mathbf{y}$ ($\mathbf{y}'(t) = V_y$).

In rectangular form, $\mathbf{r}$ is given by:  

$$ \mathbf{r} = r\big(\cos \theta\,\mathbf{i}  + \sin \theta \, \mathbf{k} \big)$$

### Affine Transforms

Below are the plots of the 2D affine transforms of a solenoid with a radius of one meter and three turns per meter.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the radius (r), turns per base unit of length (f), and the angular 'velocity' (w).
r = 1
f = 3
w = 2*np.pi*f
```

```python
# Plot the affine transforms.

x = np.linspace(-r, r, num=100)
z = np.sqrt(r**2 - x**2)

# Create figure and axes objects.

fig, ax = plt.subplots(ncols = 3, figsize=(15, 5))

# Plot the x,z transform.
ax[0].set_title(r'$x$, $z$ Transform', size = 12)
ax[0].set_xlabel('x-axis (m)')
ax[0].set_ylabel('z-axis (m)')

L = ax[0].plot(x, z, x, -z)

# Plot the y,z transform.
y = x
theta = w*y
z = r * np.sin(theta)

ax[1].set_title(r'$y$, $z$ Transform', size = 12)
ax[1].set_xlabel('y-axis (m)')
ax[1].set_ylabel('z-axis (m)')

L.extend(ax[1].plot(y, z))

# Plot the y,x transform.
x = r * np.cos(theta)

ax[2].set_title(r'$y$, $x$ Transform', size = 12)
ax[2].set_xlabel('y-axis (m)')
ax[2].set_ylabel('x-axis (m)')

L.extend(ax[2].plot(y, x))
```

# Solenoidal Path 

Suppose there exists a point moving along our solenoid.  It's position ($P$) is the terminal point of the variabe vector '$\mathbf{P}$', which is given by:  

$$ \mathbf{P}:=\mathbf{P}(t) = \mathbf{y} + \mathbf{r}$$

Recall that:  

$$ \mathbf{P}'(t) = \frac{\mathrm{d}\mathbf{P}}{\mathrm{d}t} = \mathbf{y}'(t) + \mathbf{r}'(t) = \frac{\mathrm{d}\mathbf{y}}{\mathrm{d}t} + \frac{\mathrm{d}\mathbf{r}}{\mathrm{d}t}$$

and that
$$ \mathbf{P}'(t) = \frac{\mathrm{d}\mathbf{P}}{\mathrm{d}t} =\frac{\mathbf{P}(t+\varepsilon) - \mathbf{P}(t)}{\varepsilon}$$

where $\varepsilon$ is the epsilon of the time domain (the domain of $\mathbf{P}(t)$), which is equal to the difference between two adjacent values in the domain ordered under magnitude.

The numerator of the above identity of $\mathbf{P}'(t)$ ($\mathrm{d}\mathbf{P}$) is the line element of the vector field that is congruent to the part of the solenoid traveled by our point;  More on this later.  Now, let $\mathbf{Q}$ be defined as the vector field that is given by mapping the initial point of $\mathbf{P}'(t)$ onto $P(t)$.

As for the aforementioned line element:  

$$ \mathrm{d}\mathbf{P}(t) + \mathbf{P}(t) = \mathbf{P}(t + \varepsilon)$$

Basically, via the above equation, if the initial point of the line element is equal to $P(t)$ then the terminal point of said line element, denoted as '$\mathrm{d}P(t)$', will be equal to $P(t + \varepsilon)$.  Let $\mathbf{S}$ be defined as the vector field given by this mapping.  The line element of this field is, as was mentioned earlier, $\mathrm{d}\mathbf{P}$.  Furthermore, this field is equal to $\varepsilon\mathbf{Q} = \mathbf{Q}\cdot \mathrm{d}t$.

$$ \mathbf{S} = \varepsilon\mathbf{Q} = \mathbf{Q}\cdot\mathrm{d}t$$

Since $\mathrm{d}t$ is a scalar value, dividing a vector value by it won't change it's direction, only it's magnitude.  Thus, the direction of $\mathrm{d}\mathbf{P}$ is equal to the direction of $\mathbf{P}'(t)$, which is equal to $\mathrm{d}\mathbf{P}$ divided by $\mathrm{d}t$.  This also means that $\mathbf{S}$ has the same shape as $\mathbf{Q}$, but a different size, which goes hand in hand with the proportional relationship found in the above equation.

The length of the solenoidal path of $\mathbf{P}$, denoted as '$l$', is equal to the sum of the magnitude of $\mathrm{d}\mathbf{P}$ at every point in time.

Since finding this quantity may be difficult, let's look at a 2D path first.

## Two Dimensional Path

Let's temporarily assume that the $x$, $z$ affine transform our 3D solenoid is actually a 2D structure that our point ($P$) is moving along.  The path of $P$ is thus 2D as well.  

Assuming that the time domain is continuous and $P$ made one full rotation around the the point of origin (or the y-axis if this were still a 3D space), the length of the path is equal to the circumference of a circle with a radius of $r$ meters.  

$$ l = 2\pi r$$

Let $n$ be defined as:  

$$ n:= \frac{1}{f \varepsilon}$$

If the time domain is discrete, then $\mathbf{S}$ will be congruent with an $n$-equilateral or an equilateral with $n$ edges.

### Example

For the sake of simplicity, assume that the frequency is equal to one cycle per second.  If we want $\mathbf{S}$ to be congruent with an equilateral hexagon, then the following must hold true:  

$$ n = \frac{1}{\varepsilon} = 6$$

Thus,
$$ \varepsilon = \frac{1}{6}\mathrm{s}$$

The angular velocity ($\omega$) is of course given by:  $\omega = 2\pi$, and the direction of $\mathbf{P}(t)$ is given by:  $\theta := \theta(t) = \omega \cdot t$.

The time it takes for one rotation or cycle around the point of origin is given by:  

$$ p = \frac{1}{f} = 1\mathrm{s}$$

Here, '$p$' denotes the period of the other affine transforms of $\mathbf{P}$ if it were in a 3D space.

Recall that the $x$ and $z$ complements of $\mathbf{P}$ are given by:  

$$P_x = r \cdot \cos \theta $$

and,
$$ P_z = r \cdot \sin \theta$$

Since the edges of an equilateral are all equal to one another, the perimeter of the equilateral, which is equal to $l$, is equal to the length of one edge multiplied by $n$ (the number of edges).

\begin{align*}\tag{1}
    l &= 2nr \cdot \sin \frac{\pi}{n} \\
     &= 12 \cdot \sin \frac{\pi}{6} = 6\,\mathrm{m}
\end{align*}     

```python
12*np.sin(np.pi/6)
```

```python
f = 1
w = 2 * np.pi * f
p = 1/f
n = 6

t = np.linspace(0, p, num = n + 1)
theta = w * t
P_x = x = r * np.cos(theta)
P_z = z = r * np.sin(theta)

fig, ax = plt.subplots()
ax.set_title('2D Path', size = 15)
ax.set_xlabel('x-axis (m)', size = 11)
ax.set_ylabel('z-axis (m)', size = 11)

L = ax.plot(x, z, label=r'$n = 6$; $f = 1$; $r = 1\mathrm{m}$')
ax.legend()
```

```python
(np.cos(np.pi/3), np.sin(np.pi/3))
```

The **derivative** of $\mathbf{P}(t)$ is given by:  

$$ \mathbf{P}'(t) = \frac{\mathbf{P}(t + \varepsilon) - \mathbf{P}(t)}{\varepsilon} = n \big(\mathbf{P}(t + \varepsilon) - \mathbf{P}(t) \big)$$

And, the $x$ and $z$ compliments are given by:  

$$ \mathbf{P}'(t) = P_x'(t)\,\mathbf{i} + P_z'(t)\,\mathbf{k}$$

$$ P_x'(t) = r \, \frac{\cos \omega(t + \varepsilon) - \cos \omega t}{\varepsilon} = nr\big(\cos \omega(t + \varepsilon) - \cos \omega t \big)$$

The first term of the right hand side of the above equation can be rewritten as a sine function.

$$ \cos \omega(t + \varepsilon) = \sin\big(\omega(t + \varepsilon) + 0.5\pi \big)$$

The above equation, the addition formula, and equivalence principle,  

$$ \cos \omega(t + \varepsilon) = \sin(a + b) = \sin a \, \cos b + \sin b \, \cos a$$

where, $ a = \omega t + 0.5\pi$ and $b = \omega \varepsilon$.

The above equation can be rewritten as:  

$$ \cos \omega(t + \varepsilon)  = \cos \omega t \, \cos \omega \varepsilon - \sin \omega \varepsilon \, \sin \omega t$$

Which means that:  

$$ P_x'(t) = nr\big( \cos \omega t \, \cos \omega \varepsilon - \sin \omega \varepsilon \, \sin \omega t - \cos \omega t\big)$$

$$ P_x'(t) = nr\big( \cos \omega t(\cos \omega \varepsilon - 1) - \sin \omega \varepsilon \, \sin \omega t\big)$$

$$ P_x'(t) = -6\left(0.5\,\cos\omega t + \sin \frac{\pi}{3} \, \sin \omega t \right)$$

Now for the $z$-compliment of $\mathbf{P}'(t)$.

$$ P_z'(t) = nr\big( \sin \omega(t + \varepsilon) - \sin \omega t\big)$$

Via the addition formula,  

$$ \sin \omega(t + \varepsilon) = \sin \omega t \, \cos \omega \varepsilon + \sin \omega \varepsilon \, \cos \omega t$$

Via the above two equations and substitution,  

$$ P_z'(t) = nr\big(\sin\omega t(\cos \omega \varepsilon - 1) + \sin \omega \varepsilon \, \cos \omega t \big)$$

$$ P_z'(t) = 6\left(\sin \frac{\pi}{3} \, \cos \omega t - 0.5\,\sin\omega t  \right)$$

```python
dP_x = x = -6*(0.5*np.cos(theta) + np.sin(np.pi/3)*np.sin(theta))
dP_z = z = 6*(np.sin(np.pi/3)*np.cos(theta) - 0.5*np.sin(theta))

fig, ax = plt.subplots()
ax.set_title(r'Plot of $\mathbf{P}\'(t)$', size = 14)
ax.set_xlabel('x-axis (m)', size = 11)
ax.set_ylabel('z-axis (m)', size = 11)

L = ax.plot(x, z, label=r'$n = 6$; $f = 1$; $r = 1\mathrm{m}$')
ax.legend()
```

```python
(x[0], z[0])
```

The length of each edge of the above hexagon is given by:
$$ \mathscr{l} = 2r\,\sin \frac{\pi}{n} = 2\,\sin \frac{\pi}{6}$$

Which is equal to the magnitude of every vector in $\mathbf{S}$.

\begin{equation}\tag{2}
    |\mathbf{s}| = \mathscr{l} = 2r\,\sin\frac{\pi}{n} \quad \text{where }\mathbf{s}\in\mathbf{S}
\end{equation}    

The length of the solenoid is equal to either $|\mathbf{s}|$ or $\mathscr{l}$ multiplied by the number of edges or vectors in $\mathbf{S}$.

\begin{equation}\tag{3}
    l = n\mathscr{l} = n|\mathbf{s}| = |\mathbf{S}||\mathbf{s}| = 2nr \, \sin \frac{\pi}{n}
\end{equation}    

```python
l = 12*np.sin(np.pi/6)
l
```

## Adding Depth

What if we wanted to add some depth (in the y-axis direction) to $\mathbf{S}$ and the solenoid that it is congruent to?  Recall that said solenoid is defined as the path taken by the terminal point of $\mathbf{P}$ ($P$).

Let's start by rotating every vector in $\mathbf{S}$ by $\varphi$ around the line that intersects it's initial point and is perpendicular to the $x$, $y$ plane in the direction of the positive end of the y-axis.  This means that every vector with a negative $x$-compliment will rotate in the clockwise -- or negative -- direction with respect to the $x$, $y$ Cartesian plane, while the vectors with a positive $x$-compliment will rotate in the counter-clockwise -- or positive -- direction with respect to said plane.

The absolute value of the degree of rotation ($|\varphi|$) of each vector depends on the number of turns per meter of the solenoid.  This same quantity was once called the *psuedo*-frequency ($\mathrm{pf}$) in this document.  Basically, the reciprocal of $\mathrm{pf}$, which is equal to the number of meters per turn, should be equal to the sum of the $y$-compliment of every vector in $\mathrm{S}$.

Note that this will not produce a solenoid, since the $x$, $y$ affine transform will be a triangle wave and not a sinosuedal wave, which is essential for a solenoid.  From here on out, the 3D structure that $\mathbf{S}$ is congruent to, referred to as a "solenoid", is actually something else.  Said 3D structure may, herein, be referred to as a "pseudo-solenoid".  

Furthermore, the $x$, $z$ affine transform of the structure will no longer fit a circle after rotating by a non-zero amount.  The rotational symmetry that once existed around the point of origin of the $x$, $z$ Cartesian plane is broken post-rotation.  Thus, the resulting pseudo-solenoid will not fit a cylinder.  Any future reference to a "cylinder" that is associated with the resulting pseudo-solenoid can be thought of, and referred to, as a "pseudo-cylinder".

For all $\mathbf{s} \in \mathbf{S}$, the square root of the sum of the $x$ and $y$ compliments of $\mathbf{s}$ post-rotation will be equal to the magnitude of the $x$-compliment of $\mathbf{s}$ pre-rotation, which is denoted as '$\mathbf{\grave{s}}$'.

$$ \forall \mathbf{s} \in \mathbf{S} \; : \; \sqrt{\mathbf{s}_x^2 + \mathbf{s}_y^2} = |\mathbf{\grave{s}}_x|$$

Now consider that:
$$ \sin \varphi = \frac{\mathbf{s}_y}{|\mathbf{\grave{s}}_x|}$$

As was mentioned earlier in this section,
$$ \sum_{\mathbf{s} \in \mathbf{S}} \mathbf{s}_y = \frac{1}{\mathrm{pf}}$$

The sine of $\varphi$ is also given by:  

$$ \sin\varphi = \frac{1}{4\grave{r}\mathrm{pf}}$$

Now, what about the **radius of the cylinder**, denoted as '$r$', *post*-rotation?  To find this value, we must first find the $x$-compliment of $\mathbf{s}$ *post*-rotation, which is denoted as '$\mathbf{s}_x$'.

Via the first equation of this section,  

\begin{equation}\tag{1.b}
    \mathbf{s}_x = \sqrt{|\mathbf{\grave{s}}_x|^2 - \mathbf{s}_y^2}
\end{equation}    

Via the above equation, the second equation of this section, and substitution,  

\begin{align*}
    \mathbf{s}_x &= |\mathbf{\grave{s}}_x|\sqrt{1 - \sin^2\varphi} \\
     &= |\mathbf{\grave{s}}_x|\,\cos\varphi
\end{align*}     

### Length of the Solenoid

The length ($l$) of the solenoid is unchaged by the rotation.
$$l= \sum_{\mathbf{s} \in \mathbf{S}} |\mathbf{s}| = |\mathbf{S}| |\mathbf{s}| = n \cdot |\mathbf{s}|$$

Via equation (3) in the [2D Path](#Two-Dimensional-Path) section,

$$ l = 2n\grave{r}\,\sin\frac{\pi}{n}$$

Note that the pre-rotation radius is denoted as '$r$' in the referenced equation, but is now denoted as '$\grave{r}$', while '$r$' denotes the post-rotation radius of the cylinder.

### Vector Addition and Integrals

The line integral does not vary with the number of turns per meter and subsequently $\varphi$.  The area under or over half a turn, starting with $(0,y_\mathrm{beg},0)$ and ending with $(0, y_\mathrm{end}, 0)$, of the solenoid is equal to the integral of $P_z(x)$ over $[-\grave{r}, \grave{r}]$.  Here, the $z$-compliment of $\mathbf{P}$ ($P_z$) is a function of $x=P_x$.  Note that '$\mathbf{P}$' denotes the variable vector whose path defines the solenoid pre-rotation.

This may be trivial, but:
$$ y_\mathrm{end} = y_\mathrm{beg} + \frac{1}{2\mathrm{pf}}$$

Recall that:
$$ P_x(t) = x(t) = r \, \cos \omega t$$

and,
$$ P_z(t) = r \, \sin \omega t$$

To find $P_z(x)$ we first need to get $t$ in terms of $x$, which we will do by solving for $t$ in the identity of $P_x(t)$.

$$ \cos\omega t = \frac{x}{r}$$

$$ \omega t = \arccos \frac{x}{r}$$

$$ t = \frac{1}{\omega}\,\arccos \frac{x}{r}$$

Via the above equation, the identity of $P_z(t)$, and function composition,  

\begin{align*}
    P_z(x) &= r \, \sin \arccos \frac{x}{r}\\
     &= \pm\sqrt{r^2-x^2}
\end{align*}     

Now let $P_z(x)$ be redefined as $P_z(x):=|P_z(x)|$.

The line integral is always equal to:
$$ \int_{-\grave{r}}^\grave{r} P_z(x) \, \mathrm{d}x$$

Which is given by, assuming the domain of $P_z$ is congruent with the set of real numbers:  

$$ \int_{-\grave{r}}^\grave{r} P_z(x) \, \mathrm{d}x = 0.5\pi \grave{r}^2$$

Assuming a continuous domain, the primitive integral is given by:  

$$ \int P_z(x) \, \mathrm{d}x = - \frac{2}{3}\,\frac{\sqrt{(r^2-x^2)^3}}{2x} + K = - \frac{1}{3x}\,(r^2-x^2)\,\sqrt{r^2-x^2} + K$$

In the case of our specific example, where $n=6$, the domain of $P_z$ is obviously discrete.  The integral in that case is equal to half the area of an equilateral hexagon with a radius of $\grave{r}=1$.

\begin{equation}\tag{8}
    \int_{-\grave{r}}^\grave{r} P_z(x) \, \mathrm{d}x = 3|\mathbf{s}|h
\end{equation}

Basically the integral is equal to the area of the three isosceles triangles that make up the one half of the equilateral hexagon.  Said triangles have a base that is equal to the length of each edge of the hexagon, which is equal to $|\mathbf{s}|$, and a height of $h$, which is given by:  

$$ h = \grave{r}\,\cos \frac{\pi}{n} = \cos \frac{\pi}{6}\mathrm{m}\approxeq 0.866 \mathrm{m}$$

```python
3*np.cos(np.pi/6)
```

Recall that the length of each edge of the hexagon ($\mathscr{l}$) is given by:  

$$ \mathscr{l} = |\mathbf{s}| = 2\grave{r}\,\sin\frac{\pi}{n} = 2\,\sin\frac{\pi}{6}=1\mathrm{m}$$

```python
2*np.sin(np.pi/6)
```

Which means that, in our specific case,
$$ \int_{-1\mathrm{m}}^{1\mathrm{m}} P_z(x)\,\mathrm{d}x \approxeq 3 \cdot 0.866 \mathrm{m}^2 \approxeq 2.598 \mathrm{m}^2$$

Equation (8) and related identities can be generalized to any $P_z$ with a discrete domain.  Assuming a discrete domain, the integral of $P_z(x)$ over $[-\grave{r}, \grave{r}]$ is equal to half of the area of an $n$-equilateral with a radius of $\grave{r}$.

\begin{equation}\tag{8}
    \int_{-\grave{r}}^\grave{r} P_z(x)\,\mathrm{d}x = 0.5\cdot n\cdot |\mathbf{s}|\cdot h
\end{equation}    

The above integral is equal to $0.5n$ mutliplied by the area of an isosceles triangle with a base of length $\mathscr{l}=|\mathbf{s}|$ and two legs, each with a length of $\grave{r}$.  The height of said triangle is denoted as '$h$'.

Via Pythagorean's theorem,
$$ h^2 + 0.25|\mathbf{s}|^2 = \grave{r}^2$$

$$ h = \sqrt{\grave{r}^2 - 0.25|\mathbf{s}|^2}$$

Via the above equation, equation (8), and substitution,  

\begin{equation}
    \int_{-\grave{r}}^\grave{r} P_z(x)\,\mathrm{d}x = 0.5\cdot n\cdot |\mathbf{s}|\cdot\sqrt{\grave{r}^2 -  0.25|\mathbf{s}|^2}
\end{equation}

Via the above equation, equation (2) in the [2D Path](#Two-Dimensional-Path) section, and substitution,  

\begin{align*}\tag{8}
        \int_{-\grave{r}}^\grave{r} P_z(x)\,\mathrm{d}x &= 0.5\cdot n\cdot 2\grave{r}\,\sin\frac{\pi}{n}\sqrt{\grave{r}^2 - \grave{r}^2\sin^2\frac{\pi}{n}} \\ \\
         &= 0.5\cdot n\cdot 2\grave{r}^2\,\sin\frac{\pi}{n}\sqrt{1 - \sin^2\frac{\pi}{n}} \\ \\
         &= 0.5\cdot n\cdot 2\grave{r}^2\,\sin\frac{\pi}{n}\,\cos \frac{\pi}{n} \\ \\
         &= 0.5\cdot n\cdot \grave{r}^2\,\sin\frac{2\pi}{n}
\end{align*}

### New Vector Identities

The identity of $\mathbf{P}$ and it's derivative(s) change with this rotation of $\varphi$.  To find these new identities, first let '$\mathbf{\grave{P}}$' denote the variable vector that '$\mathbf{P}$' currently denotes -- i.e. the vector *before* the rotation -- and let '$\mathbf{P}$' denote the same vector *after* said rotation.

The magnitude of $\mathrm{d}\mathbf{\grave{P}}$ does not change with the rotation.  

$$ |\mathrm{d}\mathbf{P}| = |\mathrm{d}\mathbf{\grave{P}}|$$

Assuming the same time domain pre and post rotation,  

\begin{align*}
    \mathrm{d}\mathbf{P} &= \mathbf{P}(t+\varepsilon) - \mathbf{P}(t) \\
     &= \big(P_x(t+\varepsilon) - P_x(t) \big)\mathbf{i} + \big(P_y(t+\varepsilon) - P_y(t) \big)\mathbf{j} + \big(P_z(t+\varepsilon) - P_z(t) \big)\mathbf{k} \\
     &= \mathrm{d}P_x\mathbf{i} + \mathrm{d}P_y\mathbf{j} + \mathrm{d}P_z\mathbf{k}
\end{align*}     

and,
\begin{align*}
    \mathrm{d}\mathbf{\grave{P}} &= \mathbf{\grave{P}}(t+\varepsilon) - \mathbf{\grave{P}}(t) \\
     &= \big(\grave{P}_x(t+\varepsilon) - \grave{P}_x(t) \big)\mathbf{i} + \big(\grave{P}_y(t+\varepsilon) - \grave{P}_y(t) \big)\mathbf{j} + \big(\grave{P}_z(t+\varepsilon) - \grave{P}_z(t) \big)\mathbf{k} \\
     &= \mathrm{d}\grave{P}_x\mathbf{i} + \mathrm{d}\grave{P}_y\mathbf{j} + \mathrm{d}\grave{P}_z\mathbf{k}
\end{align*} 

Via the notion that:
$$ |\mathrm{d}\mathbf{P}| = \sqrt{\mathrm{d}P_x^2 + \mathrm{d}P_y^2 + \mathrm{d}P_z^2}$$

and that,
$$ |\mathrm{d}\mathbf{\grave{P}}| = \sqrt{\mathrm{d}\grave{P}_x^2 + \mathrm{d}\grave{P}_y^2 + \mathrm{d}\grave{P}_z^2}$$

in conjunction with the first equation of this section and equivalence principle,  

$$ \sqrt{\mathrm{d}P_x^2 + \mathrm{d}P_y^2 + \mathrm{d}P_z^2} = \sqrt{\mathrm{d}\grave{P}_x^2 + \mathrm{d}\grave{P}_y^2 + \mathrm{d}\grave{P}_z^2}$$

Via dividing throughout the above equation by $\mathrm{d}t = \varepsilon$, the first equation of this section and equivalence principle, the magnitude of the derivative of $\mathbf{\grave{P}}$ with respect to time does not change with the rotation.  

$$ \left | \frac{\mathrm{d}\mathbf{P}}{\mathrm{d}t} \right | = \left| \frac{\mathrm{d}\mathbf{\grave{P}}}{\mathrm{d}t}\right|$$

The $z$-compliment also does not change;  Only the $x$ and $y$ compliments do.

Squaring another equation in this section that relates the two magnitudes post and pre rotation and then subtracting throughout by either $z$-compliment gives:  

\begin{align*}
    \mathrm{d}P_x^2 + \mathrm{d}P_y^2 &= \mathrm{d}\grave{P}_x^2 + \mathrm{d}\grave{P}_y^2 \\
    \mathrm{d}P_x^2 + \mathrm{d}P_y^2 &= \mathrm{d}\grave{P}_x^2
\end{align*}

Dividing throughout by $\mathrm{d}t^2$ gives:  

$$ \left(\frac{\mathrm{d}P_x}{\mathrm{d}t}\right)^2 + \left(\frac{\mathrm{d}P_y}{\mathrm{d}t}\right)^2 = \left(\frac{\mathrm{d}\grave{P}_x}{\mathrm{d}t}\right)^2$$

Via the above equation, notion that:
\begin{equation}\tag{9.y}
    \mathrm{d}P_y = |\mathrm{d}\grave{P}_x |\, \sin \varphi
\end{equation}    

in conjuction with substitution,
$$ \left(\frac{\mathrm{d}P_x}{\mathrm{d}t}\right)^2 + \left(\frac{\mathrm{d}\grave{P}_x}{\mathrm{d}t}\right)^2\sin^2\varphi = \left(\frac{\mathrm{d}\grave{P}_x}{\mathrm{d}t}\right)^2$$

\begin{equation}\tag{9.x}
    \frac{\mathrm{d}P_x}{\mathrm{d}t} = \frac{\mathrm{d}\grave{P}_x}{\mathrm{d}t}\cos\varphi
\end{equation}    

Multiplying throughout the above equation by $\mathrm{d}t$ and  then integrating both sides gives:  

\begin{equation}\tag{9.x}
    P_x = \grave{P}_x \, \cos\varphi + K_x
\end{equation}    

where $K_x\in\mathbb{R}$ is an arbitrary constant that must be solved for using the initial value problem (IVP) of:  
$$ P_x(0) = \grave{P}_x(0)$$

$$  \grave{P}_x(0)\,\cos\varphi + K_x = \grave{P}_x(0)$$

\begin{align*}
    K_x &= \grave{P}_x(0)(1 - \cos\varphi) \\
    &= \grave{r}\,\cos(0)(1-\cos\varphi) \\
    &= \grave{r}\cdot(1-\cos\varphi)
\end{align*}    

Thus, via the above equation, equation (9.x), and substitution,  

\begin{equation}\tag{9.x}
    P_x = \grave{P}_x \, \cos\varphi + \grave{r}\cdot(1-\cos\varphi)
\end{equation}    

To find $P_y$ in terms of $t$ we first integrate throughout equation (9.y).  

\begin{equation}\tag{9.y}
    P_y = |\grave{P}_x|\,\sin\varphi + K_y
\end{equation}    

where $K_y\in\mathbb{R}$ is an arbitrary constant we must solve for using the following IVP:  

$$ P_y(0) = 0$$

$$ \grave{P}_x(0)\,\sin\varphi + K_y = 0 $$

$$ \grave{r}\,\sin\varphi= - K_y$$

$$ K_y = - \grave{r}\,\sin\varphi$$

Thus, via the above equation, equation (9.y), and substitution,  

\begin{align*}\tag{9.y}
    P_y &= |\grave{P}_x|\,\sin\varphi - \grave{r}\,\sin\varphi \\
     &= \sin\varphi\,(|\grave{P}_x| - \grave{r})
\end{align*}

### ${x}$, $y$ Affine Transform

As was mentioned earlier, the $x$, $y$ affine transform of the structure post-rotation will be a triangle wave.  More specifically, it will be a triangle wave with peak and trough angles with a measure of $2\varphi$  and leg lengths of $\grave{r}$.  It will have an amplitude, using a peak to trough paradigm, of $r = \grave{r}\,\cos\varphi$, a peak of $\grave{r}$, and a trough of $\grave{r} - r$.  

\begin{align*}    
    \max P_y &= r = \grave{r}\,\cos\varphi \\
    \min P_y &= \grave{r} - r = \grave{r}\,(1 - \cos\varphi)
\end{align*}

The function that gives this wave, $y(x)$, is a periodic function with a period of $p_y=2\grave{r}\,\sin\varphi$.  The frequency, denoted as '$f_y$', of $y(x)$ is actually equal to the *pseudo*-frequency ($\mathrm{pf}$) of pseudo-solenoid (turns or cycles per meter).

$$ f_y = \mathrm{pf} = \frac{1}{p_y} = \frac{1}{2\grave{r}}\,\csc \varphi$$

To find $y(x)$, first let's find $P_y(P_x)$.  To start, let $P_x:=x$.  So now, we have $P_y(P_x)=P_y(x)$.  To find $P_y(x)$, we must obtain the identities of $P_x(t)$ and $P_y(t)$.

Via equation (9.x), the identity of $\grave{P}_x(t)$, and substitution,  

\begin{equation}\tag{9.x}
    P_x(t) = \grave{r}\,\cos\omega t \, \cos\varphi + \grave{r}\,(1-\cos\varphi)
\end{equation}    

And, via equation (9.y), the identity of $\grave{P}_x(t)$, and subsitution,  

\begin{align*}\tag{9.y}
    P_y(t) &= \sin\varphi\,(\grave{r}\,|\cos\omega t| - \grave{r}) \\
     &= \grave{r}\,\sin\varphi\,\big(|\cos\omega t| - 1\big)
\end{align*}
